```python
def pack_def(permutations, b):
    packs = []
    for p in permutations:
        p_units = list(p)
        p_units.extend([0] * (b-len(p)))
        new_packs = set(itertools.permutations(p_units, b))
        packs.extend(new_packs)
    return packs
```

```python
packs = pack_def(combinations, 10)
```

```python
len(packs)
```

```python
any(sum(p) > 10 for p in packs)
```

```python
def list_with_sum_generator(tope, n, suma):
    if 1 < n:
        return ([h] + tail for h in range(1, tope + 1) for tail in list_with_sum_generator(tope, n - 1, suma - h))
    else:
        return ([x] for x in xrange(1, suma + 1))
```

```python
def list_with_sum(n, max_suma):
    """Devuelve todas las listas l de enteros >0 de tamanio n tal que sum(l) <= max_suma """
    if max_suma >= n:
        tope = max_suma - (n - 1)
        return list(list_with_sum_generator(tope, n, max_suma))
    else:
        return []
```

```python
len(list_with_sum(10, 15))
```

```python
def list_with_sum2(n, max_suma):
    return list(x for x in itertools.product(range(1, max_suma + 1), repeat=n) if sum(x) <= max_suma)
```

```python
len(list_with_sum2(4, 7))
```

```python
%timeit list_with_sum2(6,10)
```

```python
def lists_with_sum_iterable(n, max_suma):
    vector = [1] * n
    i = 0
    suma = n
    
    while i < n:    
        while suma <= max_suma:
            yield tuple(vector)
            vector[0] += 1
            suma += 1
        
        i = 0
        while suma >= max_suma and i < n:
            suma -= vector[i] - 1
            vector[i] = 1
            i += 1
            
        if i < n:
            vector[i] += 1
            suma += 1      
```

```python
len(list(lists_with_sum_iterable(4, 7)))
```

```python
sum(1 for x in lists_with_sum_iterable(15, 30))
```

```python
all(len(list(lists_with_sum_iterable(n, s))) == len(list_with_sum(n, s)) for s in range(1,16) for n in range(1, s+1))
```

```python
def lists_with_sum_generator_inversed(n, max_suma):
    vector = [1] * n
    i = n - 1
    suma = n
    
    while i >= 0:    
        while suma <= max_suma:
            yield tuple(vector)
            vector[-1] += 1
            suma += 1
        
        i = n - 1
        while suma >= max_suma and i >= 0:
            suma -= vector[i] - 1
            vector[i] = 1
            i -= 1
            
        if i >= 0:
            vector[i] += 1
            suma += 1
```

```python
len(list(lists_with_sum_generator_inversed(4, 7)))
```

```python
%timeit lists_with_sum_generator_inversed(6, 10)
```
